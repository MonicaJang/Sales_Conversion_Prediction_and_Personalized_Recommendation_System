# 데이터 전처리 추가작업
- 거래취소=0(성사) 판매금액 있어야 함(0이면 안됨)
- 거래취소=1(취소) 판매금액 0이어야 함

```python
result.loc[(result['거래취소여부']==0) & (result['판매금액']==0)]

# 하기 데이터들은 오류 데이터 -> 삭제하기
```

```python
index = result[(result['거래취소여부']==0) & (result['판매금액']==0)].index
print(index)

# 위 데이터 들의 인덱스만 추출
```

```python
result1 = result.drop(index)
```

```python
result1.shape

#344266 에서 3146 개 데이터가 사라진 341120데이터만 남은걸 확인
```

```python
result1.loc[(result1['거래취소여부']==1) & (result1['판매금액']!=0)]

# 거래 취소가 됐는데 판매금액이 있음 -> 오류니까 제거
```

```python
index1 = result1[(result1['거래취소여부']==1) & (result1['판매금액']!=0)].index
print(index1)

# 위 데이터 들의 인덱스만 추출
```

```python
result2 = result1.drop(index1)
```

```python
result2.shape

# 위 오류데이터 6개 빠짐
```

```python
def month(x):
    if x >=20210601:
        return '6월'
    elif x >=20210501:
        return '5월'
    elif x >=20210401:
        return '4월'
    elif x >=20210301:
          return '3월'
    elif x >=20210201:
        return '2월'
    else:
        return '1월'


result2['월'] = result2['거래일자'].apply(month)
```

```python
result2.describe()
```

```python
# 총금액에서 수수료율을 곱한 값이 L사가 갖는 매출이 되므로
result2['L사 매출'] = result2['판매금액']*result2['수수료율']/100
```

# 전처리 완료 최종 데이터

```python
# 거래성사+취소 다 있는 전체 데이터 (거래성사 월, 매출액 컬럼 추가)
result2
```

```python
# 목표변수(Y)와 설명변수(X)를 선언  -> 새로운 X 들어올 때 Y를 예측/분류
Y = result2['거래취소여부']
X = result2[['서비스가격', '추가결제금액', '평점']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

# Over Sampling with SMOTE

```python
# 분류모델에서 비율이 깨져있으므로, 오버샘플링 시도
# SMOTE를 적용할 때는 반드시 학습 데이터 세트만 오버 샘플링을 해야함

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0)
X_train_over, Y_train_over = smote.fit_resample(X_train, Y_train)

print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트:', X_train.shape, Y_train.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 세트:', X_train_over.shape, Y_train_over.shape)
print('SMOTE 적용 후 레이블 값 분포: \n', pd.Series(Y_train_over).value_counts())
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score, roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

X_train_over, X_test, Y_train_over, Y_test = train_test_split(X, Y, test_size=0.2, random_state=156)


# 모델 생성
lgbm_model = LGBMClassifier(n_estimators=400)


# 모델 학습
lgbm_model.fit(X_train_over, Y_train_over)
```

```python
Y_train_pred_lgbm = lgbm_model.predict(X_train_over)
```

```python
print(classification_report(Y_train_over, Y_train_pred_lgbm))
```

```python
Y_test_pred_lgbm = lgbm_model.predict(X_test)
print("LGBM - Test set performance")
print(classification_report(Y_test, Y_test_pred_lgbm))
```

```python
print(classification_report(Y_test, Y_test_pred_lgbm))
```

```python
accuracy_score(Y_train_over, Y_train_pred_lgbm)
```

```python
accuracy_score(Y_test, Y_test_pred_lgbm)
```
